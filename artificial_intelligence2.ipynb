```python
# Data Library
import pandas as pd

# Library for Matrices
import numpy as np

# Machine learning Library
from sklearn.preprocessing import LabelEncoder

# Deep Learning Library
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
data = pd.DataFrame({'Mood':['Dark','Semi-Dark','Bright','Semi-Dark','Bright','Bright','Bright'],
 'Rating':['Adult','Adult','Family','Adult','Family','Adult','Adult'],
 'Class':['Dislike','Dislike','Like','Dislike','Like','Like','Like']})
```

```python
data.head(2)
```

```python
data.tail()
```

```python
data [['Mood','Rating']]
```

```python
X = pd.get_dummies(data [['Mood','Rating']])
X
# X.head(2)
```

```python
encoder = LabelEncoder()
```

```python
y = encoder.fit_transform(data['Class'])
y
```

```python
data['Class']
```

```python
encoder.classes_
```

# Model

```python
model = Sequential([
    Dense(16,input_dim=5),
    Dense(8),
    Dense(2)
    ])
```

```python
model.summary()
```

```python
np.array(X)
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X,y,epochs=3)
```
